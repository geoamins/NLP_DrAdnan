# Import Libraries

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

vader = SentimentIntensityAnalyzer()
```

```python
text = 'This was a good movie.'
vader.polarity_scores(text)                 # Compund score provides us the overall sentiment score and its value ranges from -1 to 1,
                                            #-1 being the strong negative sentiment and 1 being the strong positive sentiment.
```

```python
text = 'This was the best, most awesome movie EVER MADE!!!'
vader.polarity_scores(text)
```

```python
text = 'This was the worst film to ever disgrace the screen.'
vader.polarity_scores(text)
```
